```python
import pandas as pd
from bs4 import BeautifulSoup
import time
import requests
```

```python
# 페이지 정보 포함하도록

url = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%ED%95%9C%EA%B5%AD&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=52&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=1")
html = BeautifulSoup(url.text)

# 한페이지에 있는 뉴스 전체를 감싸는 ul안의 뉴스 한개를 감싸는 li
news = html.find("ul", class_="list_news").find_all('li',class_='bx')
len(news)
# 한 페이지 당 10개의 뉴스
```

```python
total=[]
for i in news:
    #제목 찾기
    title = i.find('a',class_='news_tit').text

    # 요약 내용 찾기
    content = i.find('a', class_='api_txt_lines dsc_txt_wrap').text

    total.append([title,content])

print(total)
len(total)
```

```python
# 데이터 프레임으로 만들기
df = pd.DataFrame(total, columns = ['제목','요약'])
df
```
