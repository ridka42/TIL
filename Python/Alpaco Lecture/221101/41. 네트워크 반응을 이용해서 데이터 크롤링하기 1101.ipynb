```python
import json
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# 구글에 업비트 검색하고 들어간후 고객센터의 프로젝트 공시

url= requests.get("https://upbit.com/service_center/disclosure")
html = BeautifulSoup(url.text)
html

```

```python
# 공지사항에서 ctrl+shift+i -> 네트워크 -> 프로젝트 공시 클릭 -> disclosure반응을 살핌 우클릭 html
# -> 새탭에서 열기 하면 제목들이 나옴 거기주소를 복사한 후 다시 크롤링
 
url = requests.get("https://project-team.upbit.com/api/v1/disclosure?region=kr&per_page=20")
url.text

# 애초에 html 형식의 문자열이 아니기 때문에 BeautifulSoup를 쓸 필요 없음
# 딕셔너리 형태와 비슷 json형태의 문자열.
```

```python
# 실제 json형식으로 바꿔주자
data = json.loads(url.text) 
data
```

```python
# 제목은 text의 키값, text는 posts의 키값, posts는 data의 키값
# 우선 data에 접근해야한다. 그 후 posts

```

```python
# for문으로 리스트를 풀어준다.

for i in data['data']['posts']:
    print(i)

# 딕셔너리 형식 -> text의 키값에 접근
```

```python
for i in data['data']['posts']:
    print(i['text'])
```

```python
# 리스트에 넣어보자

title = []
for i in data['data']['posts']:
    title.append(i['text'])

title
```
