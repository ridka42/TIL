# 2. 기본 그래프 그리기

## 2.1 y축 데이터로 그리기

* 1차원 리스트, 튜플, 시리즈 데이터를 전달하여 그린다.
* x축은 데이터의 인덱스로 자동 지정된다.


### 2.1.1 리스트로 데이터 그리기

```python


data=[1,5,3,9,7]
```

```python
plt.plot(data)
```

### 2.1.2 시리즈 데이터로 그리기

```python
import pandas as pd
```

```python
s= pd.Series([1,5,3,9,7])
s
```

```python
plt.plot(s)
```

```python
# 시리즈 데이터의 인덱스 변경하기 
ss= pd.Series([1,5,3,9,7], index=[10,20,30,40,50])
ss
```

```python
plt.plot(ss)
```

```python
sss= pd.Series([1,5,3,9,7], index=['a','b','c','d','e'])
sss
```

```python
plt.plot(sss)
```

## 2.2 x축 y축 데이터로 그리기

### 2.2.1 리스트 데이터로 그리기

```python
x=[1,5,6,9,10]
y=[1,5,3,9,7]

plt.plot(x,y)
```

### 2.2.2 데이터 프레임 데이터로 그리기

```python
sss= pd.DataFrame({'x':[1,5,6,9,10],'y':[1,5,3,9,7]})
sss
```

```python
x= sss['x']
y= sss['y']
plt.plot(x,y)
```

### 2.2.3 시리즈 데이터로 그리기

```python
s = pd.Series([1,5,3,9,7])
s
```

```python
# 시리즈만 전달해서 그리기
plt.plot(s)
plt.show()

# 인덱스가 x 축이 된다.
```

```python
# x,y 각각 전달해서 그리기
plt.plot(s.index,s.values)
# plt.plot(s.index,s) 같음
```

```python
plt.plot(s.index,s.values)
plt.show()
# 글씨 없애기
```
