# 전체 행 열과 그래프 순서에 따라 서브플롯 그리기

## 1. 라이브러리와 데이터 불러오기

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Nanum' in f.name]
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus']= False
```

```python
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
df1 = anscombe[anscombe['dataset']=="I"]
df2 = anscombe[anscombe['dataset']=="II"]
df3 = anscombe[anscombe['dataset']=="III"]
df4 = anscombe[anscombe['dataset']=="IV"]
```

## 2. 전체 행 열과 그래프 순서에 따라 서브플롯 그리기

1. figure 객체를 생성한다.
> fig= plt.figure()
2. 서브플롯을 그릴 axes 객체를 생성한다.
> ax = fig.add_subplot(전체행개수, 전체열개수, 순서)
3. axes 객체에 그래프를 그린다.
> ax.plot(x,y)
4. 축 공유하기 : 어떤 axes의 축을 공유할 것인지 지정한다.

```python
# figure 객체를 생성한다.
fig = plt.figure(figsize=(9,6))

#서브플롯을 그릴 axes 객체를 생성한다.
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
# figure 객체를 생성한다.
fig = plt.figure(figsize=(9,6))

#서브플롯을 그릴 axes 객체를 생성한다.
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# axes 객체에 그래프를 그린다.
ax1.plot(df1['x'],df1['y'],'o')
ax2.plot(df2['x'],df2['y'],'^')
ax3.plot(df3['x'],df3['y'],'*')
ax4.plot(df4['x'],df4['y'],'+')
```

```python
# figure 객체를 생성한다.
fig = plt.figure(figsize=(9,6))

#서브플롯을 그릴 axes 객체를 생성한다.
# ax1의 x축 공유하기
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2 ,sharex=ax1)
ax3 = fig.add_subplot(2,2,3, sharex=ax1)
ax4 = fig.add_subplot(2,2,4, sharex=ax1)

# axes 객체에 그래프를 그린다.
ax1.plot(df1['x'],df1['y'],'o')
ax2.plot(df2['x'],df2['y'],'^')
ax3.plot(df3['x'],df3['y'],'*')
ax4.plot(df4['x'],df4['y'],'+')

```

```python
# figure 객체를 생성한다.
fig = plt.figure(figsize=(9,6))

#서브플롯을 그릴 axes 객체를 생성한다.
# ax1의 x축 공유하기
# ax1의 y축 공유하기
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2 ,sharex=ax1, sharey=ax1)
ax3 = fig.add_subplot(2,2,3, sharex=ax1, sharey=ax1)
ax4 = fig.add_subplot(2,2,4, sharex=ax1, sharey=ax1)

# axes 객체에 그래프를 그린다.
ax1.plot(df1['x'],df1['y'],'o')
ax2.plot(df2['x'],df2['y'],'^')
ax3.plot(df3['x'],df3['y'],'*')
ax4.plot(df4['x'],df4['y'],'+')
```

```python
# 1. figure 객체를 생성한다.
fig = plt.figure(figsize=(9,6))

# 2. 서브플롯을 그릴 axes 객체를 생성한다.
# 4. ax1의 x축 공유하기
# 5. ax1의 y축 공유하기
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2 ,sharex=ax1, sharey=ax1)
ax3 = fig.add_subplot(2,2,3, sharex=ax1, sharey=ax1)
ax4 = fig.add_subplot(2,2,4, sharex=ax1, sharey=ax1)

# 3. axes 객체에 그래프를 그린다.
ax1.plot(df1['x'],df1['y'],'o')
ax2.plot(df2['x'],df2['y'],'^')
ax3.plot(df3['x'],df3['y'],'*')
ax4.plot(df4['x'],df4['y'],'+')

# 6. 틱 변경하기
ax4.set_xticks(range(1,20,1))
ax4.set_yticks(range(1,15,1))

plt.show()
```

```python
# 1. figure 객체를 생성한다.
# 10. 배경색 추가
fig = plt.figure(figsize=(9,6),facecolor='ivory')

# 2. 서브플롯을 그릴 axes 객체를 생성한다.
# 4. ax1의 x축 공유하기
# 5. ax1의 y축 공유하기
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2 ,sharex=ax1, sharey=ax1)
ax3 = fig.add_subplot(2,2,3, sharex=ax1, sharey=ax1)
ax4 = fig.add_subplot(2,2,4, sharex=ax1, sharey=ax1)

# 3. axes 객체에 그래프를 그린다.
# 7. 그래프 제목 달기
ax1.plot(df1['x'],df1['y'],'o', label='ax1')
ax2.plot(df2['x'],df2['y'],'^', label='ax2')
ax3.plot(df3['x'],df3['y'],'*', label='ax3')
ax4.plot(df4['x'],df4['y'],'+', label='ax4')

# 6. 틱 변경하기
ax4.set_xticks(range(1,20,1))
ax4.set_yticks(range(1,15,1))

# 8. 범례 표시하기
# 2는 upper left
ax1.legend(loc=2)
ax2.legend()
ax3.legend(loc='lower tight')
ax4.legend(loc=(0.5,0.5))

# 9. figure 제목 추가하기
fig.suptitle('Anscombe', size=20)

plt.show()
```

# 데이터의 크기를 비교하는 막대그래프

## 1. 라이브러리와 데이터 불러오기

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 샘플 데이터 : 어느 두 매장의 요일별 평균 매출액
df1= pd.DataFrame({'요일':['월','화','수','목','금','토','일'], '매출액':[10000,9000,11000,8000,13000,15000,14000]})
df1
```

```python
df2= pd.DataFrame({'요일':['월','화','수','목','금','토','일'], '매출액':[9000,9500,13000,7000,12000,14000,11000]})
df2
```

## 2. 막대그래프
* plt.bar(x,y)

```python
# 가로 막대 그래프

plt.bar(df1['요일'],df1['매출액'])
```

```python
# 세로 막대 그래프
plt.barh(df1['요일'],df1['매출액'])
```

## 3. 막대 폭 지정
* 가로 막대 : width
* 세로 막대 : height

```python
# 가로 막대 그래프 
plt.bar(df1['요일'],df1['매출액'], width=0.4)
# 0.8이 디폴트
```

```python
# 세로 막대 그래프 
plt.barh(df1['요일'],df1['매출액'], height=0.4)
# 0.8이 디폴트
```

## 4. 막대 색상 지정

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, color='r')
```

```python
# 막대 마다 다르게 색 지정
color = ['r','orange','y','g','b','navy','violet']
plt.bar(df1['요일'],df1['매출액'], width=0.4, color=color)
```

## 5. 막대 테두리
* edgecolor= 테두리 색상
* linewidth = 테두리 두께

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, color=color, edgecolor='gray', linewidth=2)
```

## 6. 막대 패턴 지정하기


### 6.1. 막대 패턴 지정하기
* hatch 파라미터에 기호 넣기 : / , | - + x o O . *

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, color=color, edgecolor='gray', linewidth=2, hatch='/')
```

### 6.2 패턴 밀도 지정
* hatch 기호 개수로 조절

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, color=color, edgecolor='gray', linewidth=2, hatch='///')
```

### 6.3. 막대마다 다른 패턴 지정
* 막대그래프를 객체로 받는다.
* 막대마다 다른 패턴을 지정한다.
* set_hatch(기호)

```python
bars = plt.bar(df1['요일'],df1['매출액'], width=0.4, color=color, edgecolor='gray', linewidth=2, hatch='///')
bars[0].set_hatch('.')
bars[1].set_hatch('x')
bars[2].set_hatch('+')
bars[3].set_hatch('-')
bars[4].set_hatch('*')
bars[5].set_hatch('o')
bars[6].set_hatch('|')
```

## 7. 막대 위치 지정
> align = center/edge
* 디폴트 : center
* edge로 지정하면 막대의 왼쪽 끝과 틱을 맞춘다.
* 막대의 오른쪽 끝과 틱을 맞추려면 width를 음수로 지정한다.

```python
# 왼쪽
plt.bar(df1['요일'],df1['매출액'], width=0.4, align='edge')
```

```python
# 오른쪽
plt.bar(df1['요일'],df1['매출액'], width=-0.4, align='edge')
```

### 7.1. 두 개의 막대그래프 비교하기

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4)
plt.bar(df2['요일'],df2['매출액'], width=0.4)
# 위치 조정 필요
```

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, align='edge')
plt.bar(df2['요일'],df2['매출액'], width=-0.4,align='edge')
```

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, align='edge')
plt.bar(df2['요일'],df2['매출액'], width=-0.4,align='edge')
plt.title('두 매장의 요일별 매출액 평균',size=20)
plt.grid(axis='y', ls=":")
```

```python
# 세로 막대 그래프
plt.barh(df1['요일'],df1['매출액'], height=0.4, align='edge')
plt.barh(df2['요일'],df2['매출액'], height=-0.4,align='edge')
plt.title('두 매장의 요일별 매출액 평균',size=20)
plt.grid(axis='x', ls=":")
```
