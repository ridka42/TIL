## 데이터 분석 실전

### 쇼핑몰 데이터 분석

```python
import pandas as pd
customer_master = pd.read_csv('/content/drive/MyDrive/알파코/파이썬 프로젝트/Shopping_Data/customer_master.csv')
item_master = pd.read_csv('/content/drive/MyDrive/알파코/파이썬 프로젝트/Shopping_Data/item_master.csv')
transaction_1 = pd.read_csv('/content/drive/MyDrive/알파코/파이썬 프로젝트/Shopping_Data/transaction_1.csv')
transaction_2 = pd.read_csv('/content/drive/MyDrive/알파코/파이썬 프로젝트/Shopping_Data/transaction_2.csv')
transaction_detail_1 = pd.read_csv('/content/drive/MyDrive/알파코/파이썬 프로젝트/Shopping_Data/transaction_detail_1.csv')
transaction_detail_2 = pd.read_csv('/content/drive/MyDrive/알파코/파이썬 프로젝트/Shopping_Data/transaction_detail_2.csv')
```

```python
transaction_2.head()
```

```python
len(transaction_detail_1)
```

```python
transaction_detail_1.head()
```

```python
transaction_2.head()
```

```python
len(transaction_detail_2)
```

```python
transaction_detail_2.head()
```

```python
transaction = pd.concat([transaction_1,transaction_2], ignore_index=True)
transaction
```

```python
transaction_detail = pd.concat([transaction_detail_1,transaction_detail_2], ignore_index=True)
transaction_detail
# 아이템을 여러개 샀을 경우 transaction_id가 중복 돼서 보여줌
```

```python
join_data = pd.merge(transaction_detail, transaction[['transaction_id',
                                                     'payment_date','customer_id']], 
                     on='transaction_id', how='left')
join_data
# 공통컬럼 - transaction_id
# 머지 방식 - left -> transaction_detail을 기준으로
# transaction의 price 컬럼은 제외 -> transaction_detail은 item_id로 행이 나눠지는데
# transaction은 transaction_id로 행이 나눠져서 price에 여러 아이템이 합쳐져 있기 때문에
```

```python
#  join_data(구매정보, 구매 상세정보)와 customer_master(고객 데이터), 
# item_master(아이템 데이터) 를 merge함수를 사용하여 결합
sub1 = join_data.merge(customer_master, on='customer_id', how='left')
sub1.head()
```

```python
total_data = sub1.merge(item_master, on='item_id', how='left')
total_data.head()
```

```python
total_data['price_column'] = total_data['quantity']*total_data['item_price']
total_data.head()
```

```python
total_data.isnull().sum()
```

```python
join_data.isnull().sum()
```

```python
total_data['payment_date'] = pd.to_datetime(total_data['payment_date'])
```

```python
total_data['payment_month'] = total_data['payment_date'].dt.strftime("%Y%m")
total_data[['payment_month','payment_date']].head()
```

## 머신러닝

### Regresstion

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats  # 과학용 계산 라이브러리
from sklearn.datasets import load_boston  # 사이킷런 데이타셋
from sklearn.model_selection import train_test_split # 트레인/테스트 데이터 나누기
%matplotlib inline
import warnings
warnings.filterwarnings('ignore') # 경고무시
```

```python
# boston 데이타셋 로드
boston = load_boston()

# boston 데이타셋 DataFrame 변환
df = pd.DataFrame(boston.data, columns = boston.feature_names)
df['MEDV'] = boston.target
```

```python
# input 과 target data 분리
x=df.drop(['MEDV'], axis=1) # input
y=df['MEDV'] # target
```

```python
# train과 test data 분리
# test_size=0.2 : 20퍼는 test data (보통 0.2~0.3)

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
# 학습
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# train 정확도
model.score(x_train, y_train) # 신뢰도
```

```python
# test 정확도
model.score(x_test, y_test)

# 둘의 차이가 크다면 오버피팅/언더피팅
```

```python
# 제대로 예측하는지 실제 y_test와 예측값 비교
y_test_prediction = model.predict(x_test)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_test_prediction) # 에러계산
```

## Classification

### Decision Tree

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
# iris 데이터 불러오기
iris_data = load_iris()

# x,y 데이터 셋
x = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
y = pd.DataFrame(iris_data.target, columns = ['class']) # target data = 해답
```

```python
x.head()
```

```python
# train, test 데이터 분리
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
# 학습
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# 테스트 정확도
model.score(x_test, y_test)
```

```python
# 예측 해보기
y_predict = model.predict(x_test)
y_predict
```

```python
# 실제 y_test와 비교 (오차 계산)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_predict)
```

```python
# 중요도 그래프 그리기
importance = model.feature_importances_
import matplotlib.pyplot as plot
import seaborn as sns
sns.barplot(x=x.columns, y=importance)
```
