# DataFrame 그룹 생성(groupby)

```python
os.listdir(base_src)
```

```python
abalone_src = base_src+'/abalone.data'
```

```python
abalone_df = pd.read_csv(abalone_src, header=None, sep=',', names=['sex','length','diameter','height', 
                  'whole_weight','shucked_weight','viscera_weight', 
                  'shell_weight','rings'])
abalone_df
# header는 안가져옴 ,기준으로 컬럼 나누기, names는 컬럼명
```

```python
# 데이터 shape 확인
abalone_df.shape
# 딥러닝에서 이미지(데이터수 , 가로, 세로, 채널) -> 4차원이다.
# 모양을 알기 위해
```

```python
# 데이터 결측값 확인
abalone_df.isnull().sum()
```

```python
abalone_df.isnull().sum().sum()
```

```python
# 기술 통계 확인
abalone_df.describe()
# 문자열 컬럼은 나오지 않는다. (연속형 변수(Numeric)만 확인 )
```

```python
# 전복(abalone) 성별에 따라 groupby 함수를 통해 집계
# whole_weight 값들을 성별로 나눈다. 그 후 집계 함수를 적용
grouped = abalone_df['whole_weight'].groupby(abalone_df['sex'])
grouped.sum()
# 성별 별의 whole_weight 합
```

```python
# 성별 별의 whole_weight 평균
grouped.mean()
```

```python
grouped.size()
# i 는 무게가 작다.
```

```python
# 그룹 변수가 하나가 아닌 전체 연속형 변수에 대한 집계
abalone_df.groupby(abalone_df['sex']).mean()
# 전체 데이터에 대한 성별 별 평균
```

```python
# 간단하게 표현 가능
abalone_df.groupby('sex').mean()
```

```python
# 새로운 조건에 맞는 변수 추가
abalone_df['length_bool'] = np.where(abalone_df['length']>abalone_df['length'].median(), # 중간값 보다 큰가?
                                     'length_long'# True 일 경우
                                     ,'length_short') # False 일 경우
abalone_df
# np.where은 조건에 맞는 인덱스를 반환해준다. np.where(a,b,c) 일때
# a는 조건 (b는 True 일 때의 값 / c는 False일 때의 값이다.) 
# a는 array 형태
```

```python
# 그룹변수를 2개 이상 선택해서 총계 처리
abalone_df.groupby(['sex','length_bool']).mean()
# EDA (Exploratory Data Analysis)
```

```python
# 간결하게 표현
abalone_df.groupby(['sex','length_bool'])['whole_weight'].mean()
# whole_weight에 대한 것만 보기
```

## 사용자 함수를 통한 집계

```python
# 사용자 함수를 통한 집계 --- 중요!!
import math
def avg_ceil(x,y,z):
    return math.ceil((x+y+z)/3) # ceil은 올림

abalone_df[['diameter','height','whole_weight']]
```

```python
abalone_df[['diameter','height','whole_weight']].apply(lambda x: avg_ceil(x[0],x[1],x[2]), axis=1)
# lamba x: 는 df의 한행씩 적용됨
# 한행이 컬럼순서대로 리스트 형식으로 들어감 
# 행방향으로 평균 구하기
```

```python
abalone_df[['diameter','height','whole_weight']].apply(lambda x: avg_ceil(x[0],x[1],x[2]), axis=0)
# 열 방향으로 평균
```

### 문제
1. 사용자 정의 함수 사용
2. ['diameter','height','whole_weight'] 변수 사용
3. 세 변수의 합이 1이 넘으면 True 아니면 False를 출력 후 answer 변수에 저장
4. abalone_df 에 answer 열을 추가하고 입력

```python
# 문제
def TF(x,y,z):
    if x+y+z>1:
        answer = True
    else:
        answer = False
    return answer

abalone_df['answer'] = abalone_df[['diameter','height','whole_weight']].apply(lambda x: TF(x[0],x[1],x[2]), axis=1)
```

```python
abalone_df
```

```python
# 문제
def TF(x):
    if x[0]+x[1]+x[2]>1:
        answer = True
    else:
        answer = False
    return answer

abalone_df['answer2'] = abalone_df[['diameter','height','whole_weight']].apply(lambda x: TF(x), axis=1)
abalone_df
```
