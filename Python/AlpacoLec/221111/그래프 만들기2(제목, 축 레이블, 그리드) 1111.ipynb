# 2. 데이터 프레임 만들기

```python
import pandas as pd
```

```python
df= pd.DataFrame({"월":[1,2,3,4,5],"몸무게": [80,78,75,73,70]})
df
```

## 2.1. 그래프 그리기

# 3. 제목
* plt.title('제목')
* loc : 제목 위치('left',center',right')
* pad : 타이틀과 그래프와의 간격
* color : 폰트 색상
* fontsize : 폰트 사이즈
* fontweight : 폰트 굵기 ('normal','bold','heavy','light','ultrabold','ultralight')
* fontfamily : 폰트

```python
x=df['월']
y=df['몸무게']
plt.plot(x,y,'ro-',mec='m',mfc='m',lw=3,ms=7)
plt.title('월별 몸무게 변화', loc='left', pad=10, color='r', fontsize=20, fontweight='bold', fontfamily='NanumSquareRound')

plt.show()
```

# 4. x축, y축 레이블
* plt.xlabel('레이블이름')
* plt.ylabel('레이블이름')
* loc : 위치 ('left',center',right' / 'bottom','center','top')
* labelpad : 레이블과 그래프와의 간격
* color : 폰트 색상
* fontsize : 폰트 사이즈
* fontfamily : 폰트

```python
x=df['월']
y=df['몸무게']
plt.plot(x,y,'ro-',mec='m',mfc='m',lw=3,ms=7)
plt.title('월별 몸무게 변화', loc='left', pad=10, color='r', fontsize=20, fontweight='bold', fontfamily='NanumSquareRound')
plt.xlabel('월',  labelpad=5, color = 'm', fontsize=12,fontfamily='NanumSquareRound')
plt.ylabel('몸무게(kg)', labelpad=5, color='m', fontsize=12,fontfamily='NanumSquareRound')
plt.show()
# loc 오류 - matplotlib 버전 때문인듯 하다.
```

# 3. 틱 설정
* plt.xticks(틱리스트)
* plt.yticks(틱리스트)

```python
x=df['월']
y=df['몸무게']
plt.plot(x,y,'ro-',mec='m',mfc='m',lw=3,ms=7)
# 제목
plt.title('월별 몸무게 변화', loc='left', pad=10, color='r', fontsize=20, fontweight='bold', fontfamily='NanumSquareRound')
# 축 레이블
plt.xlabel('월',  labelpad=5, color = 'm', fontsize=12,fontfamily='NanumSquareRound')
plt.ylabel('몸무게(kg)', labelpad=5, color='m', fontsize=12,fontfamily='NanumSquareRound')
# 축 범위
plt.xlim(1,12)
plt.ylim(0,100)

# 틱
plt.xticks(range(0,13,1)) 
plt.yticks(range(0,101,10))
# 그리드
plt.grid(color= 'pink', alpha=0.5, ls=':')
plt.show()
```

```python
# 틱 레이블 붙이기
xticks_label = ['','1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월','']
yticks_label = ['0kg','10kg','20kg','30kg','40kg','50kg','60kg','70kg','80kg','90kg','100kg']

x=df['월']
y=df['몸무게']
plt.plot(x,y,'ro-',mec='m',mfc='m',lw=3,ms=7)
# 제목
plt.title('월별 몸무게 변화', loc='left', pad=10, color='r', fontsize=20, fontweight='bold', fontfamily='NanumSquareRound')
# 축 레이블
plt.xlabel('월',  labelpad=5, color = 'm', fontsize=12,fontfamily='NanumSquareRound')
plt.ylabel('몸무게(kg)', labelpad=5, color='m', fontsize=12,fontfamily='NanumSquareRound')
# 축 범위
plt.xlim(1,12)
plt.ylim(0,100)

# 틱
plt.xticks(range(0,14,1), labels= xticks_label) 
plt.yticks(range(0,101,10), labels= yticks_label)
# 그리드
plt.grid(color= 'pink', alpha=0.5, ls=':')
plt.show()
```

## 3. 틱 스타일 지정
* plt.tick_params()
* diredction : 틱 위치 (in, out, inout)
* length : 틱 길이
* width : 틱 두께
* color : 틱 색상
* labelcolor : 틱 레이블 색상
* colors : 틱과 틱 레이블 색상
* pad : 틱과 레이블 사이의 거리
* labelsize : 틱 레이블 사이즈
* axis : 축 지정

```python
x=df['월']
y=df['몸무게']
plt.plot(x,y,'ro-',mec='m',mfc='m',lw=3,ms=7)
# 제목
plt.title('월별 몸무게 변화', loc='left', pad=10, color='r', fontsize=20, fontweight='bold', fontfamily='NanumSquareRound')
# 축 레이블
plt.xlabel('월',  labelpad=5, color = 'm', fontsize=12,fontfamily='NanumSquareRound')
plt.ylabel('몸무게(kg)', labelpad=5, color='m', fontsize=12,fontfamily='NanumSquareRound')
# 축 범위
plt.xlim(1,12)
plt.ylim(0,100)
# 틱
xticks_label = ['','1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월','']
yticks_label = ['0kg','10kg','20kg','30kg','40kg','50kg','60kg','70kg','80kg','90kg','100kg']

plt.xticks(range(0,14,1), labels= xticks_label) 
plt.yticks(range(0,101,10), labels= yticks_label)
# 틱 스타일
plt.tick_params(axis='x',direction='inout', length=10, width=2, colors='g', labelsize=12) 
plt.tick_params(axis='y',direction='inout', length=10, width=2, colors='b', labelsize=12) 
# 그리드
plt.grid(color= 'pink', alpha=0.5, ls=':')
plt.show()
```
