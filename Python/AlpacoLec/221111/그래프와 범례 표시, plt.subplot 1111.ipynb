## 2. 여러개의 그래프 한번에 그리기

```python
import pandas as pd

df1= pd.DataFrame({"월":[1,2,3,4,5],"몸무게": [80,78,75,73,70]})
df2= pd.DataFrame({"월":[1,2,3,4,5],"몸무게": [60,62,59,55,54]})
```

```python
df1
```

```python
df2
```

### 2.1. 여러개의 그래프 각각 그리기
* plt.show()로 그래프를 구분한다.

```python
plt.plot(df1['월'],df1['몸무게'])
plt.show()

plt.plot(df2['월'],df2['몸무게'])
plt.show()
```

### 2.2. 데이터가 다른 두 그래프 그리기

```python
plt.plot(df1['월'],df1['몸무게'])
plt.plot(df2['월'],df2['몸무게'])
plt.show()
```

```python
# x축 틱 라벨

def xtick(x):
    x=str(x)
    if x =='0' or x=='13':
        return ""
    else:
        return x+'월'
xtick_label = pd.Series(range(1,14))
xtick_label = xtick_label.apply(xtick)
xtick_label
```

```python
# y축 틱 라벨

def ytick(x):
    x=str(x)
    return x+'kg'
ytick_label = pd.Series(range(30,95,5))
ytick_label = ytick_label.apply(ytick)
ytick_label
```

```python
plt.plot(df1['월'],df1['몸무게'])
plt.plot(df2['월'],df2['몸무게'])

# 그래프 제목
plt.title('월별 몸무게 변화', size=15)

# 축 라벨
plt.xlabel('월',fontsize=12)
plt.ylabel('몸무게', fontsize=12)

# x축 틱
plt.xticks(range(1,14), labels= xtick_label)
# y축 틱
plt.yticks(range(30,95,5), labels= ytick_label)

# 그리드
plt.grid(ls=":")

plt.show()
```

### 2.3. 범례 표시하기
* 여러 개의 플롯을 동시에 그리는 경우 각 그래프가 무엇을 표시하는지 보여주기 위해 범례를 추가한다.
    1. 그래프에 레이블을 지정한다.(label=그래프명)
    2. 범레를 표시한다. plt.legend()
* 범레의 위치는 그래프에 따라 최적의 위치에 자동으로 표시된다.

```python
# 그래프 그리고 범례 달기
plt.plot(df1['월'],df1['몸무게'] , label='James')
plt.plot(df2['월'],df2['몸무게'] , label='Amy')
plt.legend()

# 그래프 제목
plt.title('월별 몸무게 변화', size=15)

# 축 라벨
plt.xlabel('월',fontsize=12)
plt.ylabel('몸무게', fontsize=12)

# x축 틱
plt.xticks(range(1,14), labels= xtick_label)
# y축 틱
plt.yticks(range(30,95,5), labels= ytick_label)

# 그리드
plt.grid(ls=":")

plt.show()
```

#### 2.3.1. 범례 위치 지정
* plt.legend(loc=위치번호)
* plt.legend(loc=위치이름)
* plt.legend(loc=(x위치,y위치))

#### 2.3.2. 범례 위치 번호 사용

```python
# 그래프 그리고 범례 달기
plt.plot(df1['월'],df1['몸무게'] , label='James')
plt.plot(df2['월'],df2['몸무게'] , label='Amy')
plt.legend(loc=4)

# 그래프 제목
plt.title('월별 몸무게 변화', size=15)

# 축 라벨
plt.xlabel('월',fontsize=12)
plt.ylabel('몸무게', fontsize=12)

# x축 틱
plt.xticks(range(1,14), labels= xtick_label)
# y축 틱
plt.yticks(range(30,95,5), labels= ytick_label)

# 그리드
plt.grid(ls=":")

plt.show()
```

#### 2.3.3. 범례 위치 이름 사용

```python
# 그래프 그리고 범례 달기
plt.plot(df1['월'],df1['몸무게'] , label='James')
plt.plot(df2['월'],df2['몸무게'] , label='Amy')
plt.legend(loc='center right')

# 그래프 제목
plt.title('월별 몸무게 변화', size=15)

# 축 라벨
plt.xlabel('월',fontsize=12)
plt.ylabel('몸무게', fontsize=12)

# x축 틱
plt.xticks(range(1,14), labels= xtick_label)
# y축 틱
plt.yticks(range(30,95,5), labels= ytick_label)

# 그리드
plt.grid(ls=":")

plt.show()
```

#### 2.3.4. 범례 x,y 위치 지정
* x,y의 길이가 1
* 그래프 밖에도 그릴 수 있다.

```python
# 그래프 그리고 범례 달기
plt.plot(df1['월'],df1['몸무게'] , label='James')
plt.plot(df2['월'],df2['몸무게'] , label='Amy')
plt.legend(loc=(1.1,0.8))

# 그래프 제목
plt.title('월별 몸무게 변화', size=15)

# 축 라벨
plt.xlabel('월',fontsize=12)
plt.ylabel('몸무게', fontsize=12)

# x축 틱
plt.xticks(range(1,14), labels= xtick_label)
# y축 틱
plt.yticks(range(30,95,5), labels= ytick_label)

# 그리드
plt.grid(ls=":")

plt.show()
```

#### 2.3.5. 범례 속성 지정
* 열 개수 : plt.legend(ncol=열개수)
* 폰트 사이즈 : plt.legend(fontsize=폰트사이즈)
* 테두리 : plt.legend(frameon=True/False)
* 음영 : plt.legend(shadow=True/False)
* 바탕색 : plt.legend(facecolor=색상)
* 테두리색 : plt.legend(edgecolor=색상)

```python
# 그래프 그리고 범례 달기
plt.plot(df1['월'],df1['몸무게'] , label='James')
plt.plot(df2['월'],df2['몸무게'] , label='Amy')
plt.legend(loc=(1.01,0.8), ncol=2, fontsize=12, frameon=True, shadow=True, facecolor='ivory', edgecolor='k')

# 그래프 제목
plt.title('월별 몸무게 변화', size=15)

# 축 라벨
plt.xlabel('월',fontsize=12)
plt.ylabel('몸무게', fontsize=12)

# x축 틱
plt.xticks(range(1,14), labels= xtick_label)
# y축 틱
plt.yticks(range(30,95,5), labels= ytick_label)

# 그리드
plt.grid(ls=":")

plt.show()
```

## 4. 서브플롯 그리기
1. 전체 그래프의 크기를 정한다.( 정하지 않으면 디폴트값)
> plt.figure(figsize=(x사이즈,y사이즈)
2. 그래프를 그려 넣을 격자를 지정한다.(전체 행 수, 전체 열 수, 그래프 순서)
> plt.subplot((전체 행 수 전체 열 수 그래프 순서)
3. 격자에 그래프를 하나씩 추가한다.

```python
plt.figure(figsize=(10,7))
plt.subplot(221)
plt.plot(df1['x'],df1['y'],'o')
plt.subplot(222)
plt.plot(df2['x'],df2['y'],'+')
plt.subplot(223)
plt.plot(df3['x'],df3['y'],'*')
plt.subplot(224)
plt.plot(df4['x'],df4['y'],'d')
```

### 4.1. 전체 그래프의 속성 지정
* figure 객체를 변수에 받는다.
* 변수.suptitle(제목) 으로 전체 그래프 제목을 표시한다.
* 변수.tight_layout() 으로 그래프의 간격, 너비를 최적화한다.

```python
fig = plt.figure(figsize=(10,7), facecolor='ivory')
fig.suptitle('Anscombe', size=20)

plt.subplot(221)
plt.plot(df1['x'],df1['y'],'o')
plt.title('ax1')

plt.subplot(222)
plt.plot(df2['x'],df2['y'],'+')
plt.title('ax2')

plt.subplot(223)
plt.plot(df3['x'],df3['y'],'*')
plt.title('ax3')

plt.subplot(224)
plt.plot(df4['x'],df4['y'],'d')
plt.title('ax4')

fig.tight_layout()
```
