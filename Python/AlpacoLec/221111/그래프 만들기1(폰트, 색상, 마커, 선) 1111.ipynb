## 5. 분포의 시각화

```python
# 1~100 까지의 랜덤 정수 1000개
import numpy as np
data = np.random.randint(1,101,1000)
data
```

### 5.1. 히스토그램

```python
plt.hist(data)
# 10 개 구간으로 나눔 -> 더 쪼개고 싶다?
```

```python
plt.hist(data, bins=100)
plt.show()
```

### 5.2. 상자수염그래프

```python
plt.boxplot(data)
# 최소값 25% 중간값(50%) 75% 최대값
```

### 5.3. 바이올린 그래프

```python
plt.violinplot(data)
# 비교적 분포가 고르기 때문에 뚱뚱하다
# 최소 최대값 선
```

## 3. 색상, 마커, 선

### 3.1. 색상
* color  = 색상
* 색생 이름을 사용한다.
* 자주 사용하는 색은 약자 사용 가능 (b,g,r,c(cyan),m(magenta), y,k(black),w)
* hex code 사용가능

```python
x=['봄','여름','가을','겨울']
y=[19,30,15,-10]
```

```python
plt.plot(x,y,color='skyblue')
```

```python
plt.plot(x,y,color='r')
```

```python
# 네이버 색상코드표
plt.plot(x,y,color='#C1B438')
```

### 3.2. 마커
* marker  : 마커종류 (* . , o v ^ < > 1 2 3 4  s p h H + x D d)
* markersize, ms : 마커 사이즈
* markeredgecolor, mec : 마커 선 색깔
* markerdedgewidth, mew : 마커 선 굵기
* markerfacecolor, mfc : 마커 내부 색깔

```python
plt.plot(x,y,color='#C1B438', marker = "*", ms='15', mec ='b', mew='3', mfc ='m')
```

### 3.3. 선
* linestyle, ls : 선스타일
> '-' solid

>> '--' dashed

> >'-.' dash-dot

>> ':' dotted

* linewidth,lw : 선 굵기

```python
plt.plot(x,y,color='#C1B438', marker = "*", ms=20,ls='-.', linewidth=3)
```

### 3.4. [색상][마커][선]

```python
# 색
plt.plot(x,y, 'k')
```

```python
# 마커
plt.plot(x,y, 'o')
```

```python
# 선
plt.plot(x,y, '--')
```

```python
# 색, 마커
plt.plot(x,y, 'k^')
```

```python
# 색,선
plt.plot(x,y, 'y--')
```

```python
# 마커, 선
plt.plot(x,y, '^:')
```

```python
# 색상, 마커, 선
plt.plot(x,y, 'r+-')
```

```python
# 색상, 마커, 선, 크기 옵션
plt.plot(x,y, 'ro-', ms=10,mec='m',mfc='pink')
```
