```python
import re
f= open("코로나.txt","rt")
covid = f.read()
f.close()
print(covid) # 코로나를 검색했을 때 나오는 결과값들(많은 문자열)
```

```python
re.findall("[가-힣]", covid)

# 한글찾기 -> 한글자씩
```

```python
re.findall("[가-힣]+", covid)

# +를 붙이면 한개 이상으로 이뤄진 한글을 찾고싶다.
# 한글찾기 -> 한단어씩
```

```python
re.findall("[가-힣]{2}", covid)

# 두글자로 이뤄진 한글 찾기
```

```python
re.findall("[가-힣]{3,}", covid)

# 세개 이상으로 이뤄진 한글
```

```python
re.findall("[a-zA-Z]", covid)

# 영어 한글자
```

```python
re.findall("[a-zA-Z0-9]+", covid)

# 영어, 숫자 , 한개 이상으로 이뤄진 글자
```

```python
re.findall("[\d]+", covid)

# 한개 이상 숫자
```

```python
re.findall("[\D]+", covid)

# 숫자 아닌것
```

```python
# 코로나19 만 뽑고 싶을 때
re.findall("\w+19", covid)

# 19로 끝나는 한개 이상의 글자 혹은 숫자
# 2019 등 다른 것도 포함
```

```python
re.findall("[\w+19]", covid)
# 문자열이 한개 이상 , 1과 9
```

```python
# 2019 등 코로나19 외의 단어를 빼고싶으면
re.findall("코\w+19", covid)
```
