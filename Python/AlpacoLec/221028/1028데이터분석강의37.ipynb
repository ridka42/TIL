```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
# 링크의 페이지를 요청해서 받아온 데이터를 url에 저장
url = requests.get("https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EB%A1%9C%EB%98%90")
url

# 200 : 페이지 요청에 큰 문제가 없었다는 뜻
# 400번대 : 존재하지 않는 페이지
# 500번대 : 존재는 하지만 볼 수 없는 페이지(로그인처리,비공개사이트)
```

```python
# 가져온 정보
url.text
```

```python
# 문자열을 HTML로 변환
html = bs(url.text)
html
```

```python
# 1038회 라는 글씨를 가져오고 싶다.

html.find('span',class_='f_red') # 태그, 속성값

#<태그>text<태그>
```

```python
# 1038만 가져오고 싶다.

current =int(html.find('span',class_='f_red').text.replace("회",""))
current
```

```python
# 로또번호들 가져오기
# div 태그 안에 span으로 번호 하나씩

lotto = html.find("div",class_="lottonum").find_all('span')
lotto
```
