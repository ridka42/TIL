```python
from sklearn.datasets import load_diabetes # 당뇨병 데이터
import seaborn as sns # 데이터분석 연습용
import pandas as pd # 데이터분석
import scipy.stats as stats # 상관계수를 분석하기 위해
```

```python
diabetes = load_diabetes() #dic
diabetes.keys()
```

```python
data,target = diabetes['data'],diabetes['target']
data
```

```python
target 
# 당뇨병 진행도
```

```python
data.shape # 442행 10열
```

```python
df = pd.DataFrame(data, columns=diabetes['feature_names'])
df
# diabetes의 data 키의 값들을 데이터프레임으로 만듦
# 데이터프레임의 컬럼명은 diabetes의 feature_names라는 키의 값들
# 특성명들이 들어있다.
```

```python
df['target'] = target
df
```

```python
# 어떤 특성이 타겟특성과 상관관계가 높을까

x =df['age']
y =df['target']

stats.pearsonr(x,y)
# 상관계수, p-value
# 상관계수 낮음
# p-value : 상관계수 없을 확률이 거의 0
# 상관계수가 있다고 해석 가능
# 있지만 낮다
```

```python

x =df['sex']
y =df['target']

stats.pearsonr(x,y)
# 상관게수 0에 가까움 -> 거의 없다
# p-value가 0.05보다 작으면 상관관계가 있다고 해석하는데
# 지금은 더 크기때문에 상관관계가 있다고 믿기 힘들다.
```

```python

x =df['bmi']
y =df['target']

stats.pearsonr(x,y)
# 상관관계가 꽤 높다.
# p-value 매우 작음 -> 상관관계가 무조건 있다.
```

```python
cor = df.corr()
cor
# 상관계수 표 한번에 구하기
```
