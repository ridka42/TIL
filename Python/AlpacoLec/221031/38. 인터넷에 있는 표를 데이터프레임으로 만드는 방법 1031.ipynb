```python
import requests
from bs4 import BeautifulSoup
import time 
import pandas as pd
```

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1")
# 주소를 가져올 때 최대한 많은 정보를 담아서 가져온다.
# ex) sosok과 page

html = BeautifulSoup(url.text)

table = html.find('table',class_ = 'type_2')
pd.read_html(str(table))
# 리스트 안에 들어있는 데이터
```

```python
table = pd.read_html(str(table))[0]
table
# 표의 실선이 누락값으로 표시
```

```python
table = table[table['종목명'].notnull()]
table

# 결측값이 종목명에는 있으면 안되기 때문에 지워줌
```

```python
# 필요없는 칼럼 지워줌

del table['N']
del table['토론실']
table
```

```python
# 한번에 모든 코스피 페이지 가져오기
# 맨뒤 버튼인 a태그는 클래스가 없음 -> 그 상위 태그로 간다.

html.find('td',class_="pgRR").find('a')
```

```python
# 마지막 페이지 숫자만 뽑기
kospi_page = html.find('td',class_="pgRR").find('a')['href'].split("=")[-1]
```

```python
# 모든 페이지 데이터 가져오기 -> for 문 돌리기
total=[]

for p in range(1,int(kospi_page)+1):
    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={p}")
    html = BeautifulSoup(url.text)

    table = html.find('table', class_='type_2')
    table = pd.read_html(str(table))[0]
    table = table[table['종목명'].notnull()]

    del table['N']
    del table['토론실']

    total.append(table)
    time.sleep(1) # 딜레이. 공격으로 인식하지 않게
    print("{}페이지 저장완료.".format(len(total)))
```

```python

len(total)
# 3페이지의 데이터
total[2]
```

```python
kospi = pd.concat(total, ignore_index = True)
kospi

# 행번호 정리해서 전부 합치기(세로로)
```

```python
kospi.to_excel("kospi.xlsx")

#엑셀에 넣기
```

코스닥까지 합쳐서 엑셀만들기

```python
total=[]

for s in range(0,2):
    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok={s}&page=1")
    html = BeautifulSoup(url.text)
    page = int(html.find('td',class_='pgRR').find('a')['href'].split("=")[-1])
    if s==0:
        message="코스피"
    else:
        message="코스닥"
    print(message)

    for p in range(1,page+1):
        url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok={s}&page={p}")
        html = BeautifulSoup(url.text)
        table = html.find('table', class_='type_2')
        table = pd.read_html(str(table))[0]
        table = table[table['종목명'].notnull()]
        del table['N']
        del table['토론실']

        total.append(table)
        time.sleep(1)
        print(f"{p}페이지 저장 완료")
```

```python
kospi_kosdaq = pd.concat(total, ignore_index=True)
kospi_kosdaq
```

```python
kospi_kosdaq.to_excel("kospi_kosdaq.xlsx")
```
