```python
import pandas as pd

df= pd.read_excel("성적 처리.xlsx", sheet_name="Sheet1")
df
```

```python
# 과목별 평균
df.groupby("반").mean()
```

```python
# 총합이라는 새로운 열을 만드는 방법

df['총합'] = df['국어']+df['영어']+df['수학']+df['사회']+df['과학']
df
```

```python
# 더 간단한 방법
df.iloc[:,2:]
```

```python
# 각 행 별로 합 구하기 -> 행방향 합치기
df['총합'] = df.iloc[:,2:].sum(axis=1)
df
```

```python
# 학생별 평균 구하기

df['평균'] = df['총합']/5
df
# 홍길동과 다니엘은 동점
```

```python
# 평균을 순위 매기기
df['순위'] = df['평균'].rank(method='average', ascending=False)
df
# ascending True로 하면 점수 낮은 사람이 높은 등수
# method = average는 3.5등 두명 4등 존재 안함
```

```python
df['순위'] = df['평균'].rank(method='min', ascending=False)
df
# 3등 두명 4등 없음
```

```python
df['순위'] = df['평균'].rank(method='max', ascending=False)
df
# 4등 두명 3등 없음
```

```python
df['순위'] = df['평균'].rank(method='first', ascending=False)
df
# 3등 한 명 4등 한 명
```

```python
df['순위'] = df['평균'].rank(method='dense', ascending=False)
df
# 3등 두명 4등 있음
```
