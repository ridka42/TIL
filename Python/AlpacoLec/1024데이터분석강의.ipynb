ex2는 ex1에서 제공하는 모든 함수와 변수들을 사용할 수 있음

```python
class Bank:
  def deposit(self, a):
    self.money = a

person1 = Bank()
person1.deposit(5000)

person2 = Bank()
person2.deposit(3000)

print(person1.money)
print(person2.money)

# self는 닉네임이 들어갈 빈칸 같은것
```

```python
person3 = Bank()
print(person3.money)

# 값을 안넣으면 에러
```

```python
class Bank:
  def __init__(self):
    print("가입을 축하합니다!")
    self.money = 0

  def deposit(self, a):
    self.money += a

person3 = Bank()
print(person3.money)

# 값을 안넣으면 0이 나오게 -> init (초기화) 객체를 생성할 때 자동으로 한번 실행됨
```

```python
# bank2 는 bank의 모든 함수들을 상속받는다.

class Bank2(Bank):
  pass

person4 = Bank2()
```

금액을 출금하는 기능을 추가해보세요.

```python
class Bank3:
  def __init__(self):
    print("가입추카")
    self.money = 0
 

  def deposit(self, a):
    self.money += a
  

  def withdraw(self, a):
    self.money -= a
```

```python
person5 = Bank3()
```

# 15. 예외처리

에러가 발생했을 때, 에러를 무시하는 기술

예외처리를 해야하는 경우
1. 내 코드에 아무런 문제가 없을 때
2. 어디론가 이동할 때

Try:

 맛집 블로그 크롤링

except 에러이름:

 특정 에러가 났을 때만 다음 맛집으로 넘어가라

finally:
  창 닫기

 에러가 나든 안나든 try가 끝나면 무조건 finally 실행

에러 일부러 발생시키기

raise 에러이름

```python
try:
  print("Hello")
  print(me)
  print("Bye")
except:
  print("Error!!")

  # me에서 에러가 났기 때문에 bye는 안나옴
```

```python
try:
  print("Hello")
  print(me)
  print("Bye")
except:
  print("Bye")
finally:
  print("Error fin")
```

```python
try:
  if 2 == True:
    raise NameError
except:
  pass
```

# 18. 수치 계산 라이브러리 - Numpy

다차원의 배열 자료구조를 다룰 수 있으며, 벡터와 행렬을 사용하는 선형대수 계산에 주로 사용

```python
import numpy as np

# np.array: 행렬 변환

array1 = np.array([1,2,3])
print(type(array1))
print(array1.shape) # 크기 3, -> 1차원의 데이터가 3개 있다.
```

```python
# [1,2,3]+3 -> 에러남
# 행렬 -> 병렬계산 가능

np.array([1,2,3])+3
```

```python
array2 = np.array([[1,2,3],[4,5,6]])
print(type(array2))
print(array2.shape) # (2,3) -> 2행3열 이다.
```

```python
# ndim: 차원 확인

print(array1.ndim) # [1,2,3] 1차원 리스트
print(array2.ndim) # [[1,2,3],[4,5,6]] 2차원 리스트 (리스트 안의 리스트)
```

```python
# dtype: 데이터 타입
array1 = np.array([1,2,3])
print(array1, array1.dtype) # 정수

array2 = np.array([1,2,3,'test'])
print(array2, array2.dtype) # 문자열

array3 = np.array([1,2,3,4.0])
print(array3, array3.dtype) # 실수

array4 = array1.astype('float64')
print(array4, array4.dtype) # 정수를 실수로 변환

array5 = np.array([1,1.2,2,3.5]).astype('int64')
print(array5, array5.dtype) # 실수를 정수로 변환
```

```python
# np.arange: 행렬 편리하게 생성하기

seq_array = np.arange(10)
print(seq_array)
print(seq_array.dtype, seq_array.shape)
```

```python
# zeros: 0으로 채워진 행렬 만들기

zero_array = np.zeros((3,2),dtype='int32')
print(zero_array)
print(zero_array.dtype,zero_array.shape)
```

```python
# ones: 1로 채워진 행렬 만들기

one_array = np.ones((3,2),dtype='int32')
print(one_array)
print(one_array.dtype,one_array.shape)
```

```python
# reshape: 차원바꾸기

array1 = np.arange(10)
print('array1:\n',array1)

array2 = array1.reshape(2,5) # 개수가 같아야함
print('array2:\n', array2)
```

```python
array1 = np.arange(8)
array3 = array1.reshape(2,2,2)
print('array3:\n',array3.tolist()) # 리스트안에 리스트안에 리스트
# .tolist()-> 행렬을 리스트로 변환
```

```python
# 행렬 정렬: sort()

org_arr = np.array([3,6,2,4])
print('원본:', org_arr)
sort_arr1 = np.sort(org_arr)
print('sort()호출 후 반환된 행렬:',sort_arr1)
print('sort()호출 후 원본 행렬:',org_arr)
# 원본에 저장은 안됨 // 새로만든데 저장됨
```

```python
org_arr = np.array([3,6,2,4])
sort_arr2 = org_arr.sort()
print('sort()호출 후 반환된 행렬:',sort_arr2)
print('sort()호출 후 원본 행렬:',org_arr)
# 원본에 저장됨 // 새로만든데 저장안됨
```

```python
org_arr = np.array([3,6,2,4])
sort_arr3 = np.sort(org_arr)[::-1]
print('내림차순 정렬:',sort_arr3)
```

```python
array2 = np.array([[8,12],[7,1]])
print(array2)

sort_arr = np.sort(array2,axis=0)
print('행방향 정렬:\n', sort_arr) # 세로로 정렬

sort_arr = np.sort(array2,axis=1)
print('열방향 정렬:\n', sort_arr) # 가로로 정렬
```

```python
# argsort: 정렬된 데이터의 인덱스를 가져옴

org_arr = np.array([3,6,2,4])
sort_index = np.argsort(org_arr)
print('정렬시 원본행렬의 인덱스:',sort_index)
```

```python
org_arr = np.array([3,6,2,4])[::-1]
sort_index = np.argsort(org_arr)
print('내림차순 정렬시 원본행렬의 인덱스:',sort_index)
```

```python
# 행렬 내적 (행렬곱)

A = np.array([[1,2,3],[4,5,6]])
B = np.array([[7,8],[9,10],[11,12]])
dot_product = np.dot(A,B)
print('행렬 내적 결과:\n',dot_product)
```

```python
# 전치 행렬

A = np.array([[1,2,3],[4,5,6]])
transpose_A = np.transpose(A)
print("A의 전치행렬:\n",transpose_A)
```
