# 1. rcParams
* rcParams : Runtime Configuration Parameters
* 그래프를 구성하는 공동 속성을 지정한다.

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Nanum' in f.name]
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'
# - 기호 깨짐 방지
plt.rcParams['axes.unicode_minus']= False
```

```python
# 다른 rcParams
plt.rcParams
```

## 1. 라이브러리와 스타일 지정

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'
# - 기호 깨짐 방지
plt.rcParams['axes.unicode_minus']= False
```

```python
# 그리드 스타일 : 라인 스타일'--', 투명도 0.3
plt.rcParams['grid.linestyle']='--'
plt.rcParams['grid.alpha']= 0.3

# 폰트 사이즈 12
plt.rcParams['font.size']=12
```

## 2. 그래프 영역 채우기

### 2.1. 가로방향으로 채우기
* plt.fill_between(x슬라이싱, y슬라이싱)
* 슬라이싱 범위가 같아야 한다.

```python
x=[1,2,3,4]
y=[1,4,3,6]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,3,6]
plt.plot(x,y)
plt.fill_between(x[1:3],y[1:3], color='skyblue',alpha=0.5,hatch='/')
plt.show()
# x 의 인덱스가 1부터 3이면 y도 같아야함
```

### 2.2. 세로방향으로 채우기
* plt.fill_between(y슬라이싱, x슬라이싱)
* 슬라이싱 범위가 같아야 한다.

```python
x=[1,2,3,4]
y=[1,4,3,6]
plt.plot(x,y)

plt.fill_between(y[:2],x[:2], color='skyblue',alpha=0.5,hatch='/')

plt.show()
```

### 2.3. 두 그래프 사이의 영역 채우기
* plt.full_between(x슬라이싱,y1슬라이싱, y2슬라이싱)

```python
x=[1,2,3,4]
y1=[1,3,4,6]
y2=[2,4,7,9]

plt.plot(x,y1)
plt.plot(x,y2)

plt.show()
```

```python
x=[1,2,3,4]
y1=[1,3,4,6]
y2=[2,4,7,9]

plt.plot(x,y1)
plt.plot(x,y2)

plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='darkorange', alpha=0.5)

plt.show()
```

### 2.4. 다각형 채우기
* plt.fill([x축의 좌표들],[y축의 좌표들])

```python
x=[1,2,3,4]
y=[1,4,3,6]
plt.plot(x,y)
plt.grid()

plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,3,6]
plt.plot(x,y)
plt.grid()

plt.fill([1.75,2.25,2.25,1.75],[3.5,3.5,4.5,4.5], alpha=0.2)
plt.fill([2.75,3,3.25],[2.5,3.5,2.5], color='pink', alpha=0.3)
# 왼쪽에서 오른쪽으로 가는 순 으로 좌표를 적는다.

plt.show()
```
