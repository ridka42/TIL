```python
import pandas as pd
import matplotlib.pyplot as plt

# 그래프에 한글폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

# - 기호 깨짐 방지
plt.rcParams['axes.unicode_minus']= False

# 폰트 사이즈 12
plt.rcParams['font.size']=12
```

```python
# 샘플 데이터 : 어떤 학생의 나이에 따른 키와 몸무게의 변화

age = [13,14,15,16,17]
height = [160,165,170,173,177]
weight = [80, 85, 83, 78, 73]
```

## 2. 두가지 정보를 하나의 그래프에 그리기
* 축을 분리하기 위해 객체지향으로 그린다.
* fig, ax = plt.subplots()

```python
fig , ax = plt.subplots()
ax.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')
ax.plot(age,weight,color='darkred', marker='o', ls='-.',label='weight')
plt.legend()
plt.show()
# 키와 몸무게에 대한 y 축을 각각 만들려면?
```

### 2.1. 2중 y축 만들기
* x 축을 공유하는 새로운 axes 객체를 만든다.

    axes객체.twinx()

```python
fig , ax1 = plt.subplots()

ax1.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')
# x축을 공유하는 새로운 axes객체를 생성한 것
ax2 = ax1.twinx()

plt.show()
# 오른쪽에 새로운 y축이 생김 
```

```python
fig , ax1 = plt.subplots()

ax1.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')

ax2 = ax1.twinx()
# 두번째 그래프 그리기
ax2.plot(age,weight,color='darkred', marker='o', ls='-.',label='weight')

plt.show()
```

### 2.2. 축 레이블 표시하기
* axes객체.set_xlabel(x레이블)
* axes객체.set_ylabel(y레이블)

```python
fig , ax1 = plt.subplots()

ax1.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age,weight,color='darkred', marker='o', ls='-.',label='weight')

# 레이블 표시
ax1.set_xlabel('나이')
ax1.set_ylabel('키')
ax2.set_ylabel('몸무게')

plt.show()
```

### 2.3. y축 범위 지정

```python
fig , ax1 = plt.subplots()

ax1.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age,weight,color='darkred', marker='o', ls='-.',label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키')
ax2.set_ylabel('몸무게')

# y축 범위 지정
ax1.set_ylim(150,180)
ax2.set_ylim(60,90)

plt.show()
```

### 2.4. y축 눈금

```python
fig , ax1 = plt.subplots()

ax1.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age,weight,color='darkred', marker='o', ls='-.',label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키')
ax2.set_ylabel('몸무게')

ax1.set_ylim(150,180)
ax2.set_ylim(60,90)

# 데이터 값만 눈금으로 표시
ax1.set_yticks(height)
ax2.set_yticks(weight)

plt.show()
```

### 2.5. 눈금 스타일

```python
fig , ax1 = plt.subplots()

ax1.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age,weight,color='darkred', marker='o', ls='-.',label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키')
ax2.set_ylabel('몸무게')

ax1.set_ylim(150,180)
ax2.set_ylim(60,90)


ax1.set_yticks(height)
ax2.set_yticks(weight)

# 색상 통일
ax1.tick_params(axis='y', colors='skyblue')
ax2.tick_params(axis='y', colors='darkred')

plt.show()
```

### 2.6. 범례 표시
* axes 객체별로 legend 메소드 호출

```python
fig , ax1 = plt.subplots()

ax1.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age,weight,color='darkred', marker='o', ls='-.',label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키')
ax2.set_ylabel('몸무게')

ax1.set_ylim(150,180)
ax2.set_ylim(60,90)


ax1.set_yticks(height)
ax2.set_yticks(weight)

ax1.tick_params(axis='y', colors='skyblue')
ax2.tick_params(axis='y', colors='darkred')

# 범례 표시
ax1.legend()
ax2.legend()

plt.show()
```

### 2.7. 그리드 표시

```python
fig , ax1 = plt.subplots()

ax1.bar(age,height,color='skyblue', width=0.7, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age,weight,color='darkred', marker='o', ls='-.',label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키')
ax2.set_ylabel('몸무게')

ax1.set_ylim(150,180)
ax2.set_ylim(60,90)


ax1.set_yticks(height)
ax2.set_yticks(weight)

ax1.tick_params(axis='y', colors='skyblue')
ax2.tick_params(axis='y', colors='darkred')

ax1.legend()
ax2.legend()

# 그리드 표시
ax1.grid(axis='y',ls='--', color='skyblue')
ax2.grid(axis='y',ls='--', color='pink')
plt.show()
```

# Seaborn 막대그래프

## 1. 샘플 데이터

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
# 결측치 없음
```

## 2. 요일별 팁 평균

### 2.1. matplotlib으로 시각화

#### 2.1.1. 데이터 가공
* 요일별 팁 평균 계산
* 그룹핑 : 데이터프레임.groupby(그룹기준컬럼)[통계적용컬럼].통계함수

```python
day_tip_mean = tips.groupby('day')['tip'].mean()
day_tip_mean
```

#### 2.1.2. 데이터 시각화
* plt.plot(x,y)

```python
plt.bar(day_tip_mean.index,day_tip_mean)
plt.xlabel('day')
plt.ylabel('tip')
plt.title('요일별 팁 평균')
plt.show()
```

### 2.2. seaborn으로 그리기
* sns.barplot(data=데이터프레임명, x=x축컬럼, y=y축컬럼)
* x축 데이터로 그루핑한 y축 데이터의 평균값을 계산하여 그래프를 그려준다.
* 신뢰구간을 함께 표시한다.

```python
sns.barplot(data= tips, x= 'day', y='tip')
```

```python
sns.barplot(data= tips, x= 'day', y='tip', ci=None)
plt.title('요일별 팁 평균',size=15)
plt.show()
```

## 3. 요일별 팁 합계

### 3.1. matplotlib으로 그리기

```python
day_tips_sum = tips.groupby('day')['tip'].sum()
day_tips_sum
```

```python
plt.bar(day_tips_sum.index,day_tips_sum)

plt.xlabel('day')
plt.ylabel('tip')

plt.title('요일별 팁 합계', size=15)

plt.show()
```

### 3.2. seaborn으로 그리기
* estimator=통계함수

```python
sns.barplot(data=tips, x='day',y='tip',ci=None, estimator=sum)

plt.title('요일별 팁 합계', size=15)

plt.show()
```

## 4. 요일별 팁 합계를 흡연여부로 비교
* hue= y를 그룹핑할 컬럼
* hye 색상 변경 : palette= {구분:색상}

```python
sns.barplot(data=tips, x='day', y='tip',ci=None, 
            estimator=sum, hue='smoker')
plt.title('요일별 팁 합계', size=15)
plt.show()
```

```python
# 색 변경

sns.barplot(data=tips, x='day', y='tip',ci=None, 
            estimator=sum, hue='smoker', palette={'Yes':'gray','No':'skyblue'})

plt.title('요일별 팁 합계', size=15)
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip',ci=None, 
            estimator=sum, hue='smoker', palette={'Yes':'gray','No':'skyblue'})

plt.title('요일별 팁 합계', size=15)

# 그리드 추가
plt.grid(axis='y',ls=':')

plt.show()
```
