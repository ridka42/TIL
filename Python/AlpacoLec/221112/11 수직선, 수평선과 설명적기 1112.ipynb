```python
import pandas as pd
import matplotlib.pyplot as plt

# 그래프에 한글폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

# - 기호 깨짐 방지
plt.rcParams['axes.unicode_minus']= False

# 그리드 스타일 : 라인 스타일'--', 투명도 0.3
plt.rcParams['grid.linestyle']='--'
plt.rcParams['grid.alpha']= 0.3

# 폰트 사이즈 12
plt.rcParams['font.size']=12
```

## 2. 수평선, 수직선 그리기

### 2.1. 수평선 그리기
* plt.axhline(y좌표, x축 시작 위치, x축 끝 위치)

    수평선의 길이가 1이라고 했을 때 x축 시작 위치, x축 끝 위치를 지정한다. 따로 지정하지 않으면 x축 전범위
* plt.hlines(y, x축 시작 좌표, x축 끝 좌표)

     x 좌표를 지정해야만 한다.

```python
plt.plot([1,2,3,4],'ko')

plt.axhline(2, 0, 0.34, alpha=0.5, ls=':')

plt.hlines(3,0,2, alpha=0.5, color='darkorange')

plt.show()
```

### 2.2. 수직선 그리기
* plt.axvline(x좌표, y축 시작 위치, y축 끝 위치)

    수평선의 길이가 1이라고 했을 때 y축 시작 위치, y축 끝 위치를 지정한다. 따로 지정하지 않으면 y축 전범위
* plt.vlines(x, y축 시작 좌표, y축 끝 좌표)

     y 좌표를 지정해야만 한다.

```python
plt.plot([1,2,3,4],'ko')

plt.axvline(3, ls='--', color='r')
plt.vlines(2,0,3, color='b')
plt.show()
```

# 그래프에 설명 적기

## 1. 텍스트 추가하기
*  plt.text(x좌표, y좌표, 텍스트)
* rotation=회전각도
* ha : horizontal alignment
* va : vertical alignment
* 텍스트 상자
> bbox = {'boxstyle' : 상자스타일, 'fc' : facecolor, 'ec' : edgecolor, ...}<br>boxstyle : 'round/square'

```python
plt.plot([1,2,3,4],'ko')
plt.text(2.1,3,'(x=2, y=3)')

plt.show()
```

```python
# 박스의 좌표 기준 정하기
plt.plot([1,2,3,4],'ko')
plt.text(2,3,'(x=2, y=3)',ha='right',va='top')

plt.show()
```

```python
# 박스 꾸미기

plt.plot([1,2,3,4],'ko')
plt.text(2.1,3,'(x=2, y=3)', fontsize = 12, rotation=30,
         bbox={'boxstyle':'round', 'fc':'skyblue', 'ec':'b', 'alpha':0.3})

plt.show()
```

```python
plt.plot([1,2,3,4],'ko')
plt.text(2.1,3,'(x=2, y=3)', fontsize = 12, rotation=30,
         bbox={'boxstyle':'round', 'fc':'skyblue', 'ec':'b', 'alpha':0.3})
# 수직선 그어보기
plt.axvline(2,color='pink',lw=0.5)
# 수평선 그어보기
plt.axhline(3,color='y', alpha=0.5, lw=0.5)
plt.show()
```

```python
plt.plot([1,2,3,4],'ko')
plt.text(2.1,3,'(x=2, y=3)', fontsize = 12, rotation=30,
         bbox={'boxstyle':'round', 'fc':'skyblue', 'ec':'b', 'alpha':0.3})
plt.axvline(2,color='pink',lw=0.5)
plt.axhline(3,color='y', alpha=0.5, lw=0.5)

# 특정 마커 바꾸기
plt.plot(2,3,'bo')
plt.show()
```

## 2. 화살표와 텍스트 추가하기
* plt.annotate('텍스트',xy=(화살표의 x좌표, 화살표의 y좌표), xytext(텍스트의 x좌표,텍스트의 y좌표), arrowprops=화살표속성(딕셔너리))
* 화살표 속성
    
    width : 화살표 넓이

    headwidth : 화살표 머리 넓이

    headlength : 화살표 머리 길이

    shrink : 화살표와 점과의 거리

```python
plt.plot([1,2,3,4],'ko')
plt.axvline(2,color='pink',lw=0.5)
plt.axhline(3,color='y', alpha=0.5, lw=0.5)
plt.plot(2,3,'bo')

# 텍스트 추가
plt.annotate('(x=1, y=2)', xy=(1,2), xytext=(1.2,2.2))

plt.show()
```

```python
plt.plot([1,2,3,4],'ko')
plt.axvline(2,color='pink',lw=0.5)
plt.axhline(3,color='y', alpha=0.5, lw=0.5)
plt.plot(2,3,'bo')

# 화살표 + 텍스트 추가
plt.annotate('(x=1, y=2)', xy=(1,2), xytext=(1.5,2.5), arrowprops={})

plt.show()
```

```python
plt.plot([1,2,3,4],'ko')
plt.axvline(2,color='pink',lw=0.5)
plt.axhline(3,color='y', alpha=0.5, lw=0.5)
plt.plot(2,3,'bo')

# 화살표 + 텍스트 꾸미기
plt.annotate('(x=1, y=2)', xy=(1,2), xytext=(1.5,2.5), 
             arrowprops={'width':1, 'headwidth':8, 'headlength':8, 
                         'shrink':0.1, 'fc':'r'}, 
             fontsize=12, color='r')

plt.show()
```
