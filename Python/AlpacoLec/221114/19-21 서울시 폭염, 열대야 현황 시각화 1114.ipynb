```python
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus']= False
```

### 1.1 데이터프레임 생성

```python
# 파일 목록 저장하기
file_list = os.listdir('./drive/MyDrive/Colab Notebooks/알파코 파이썬 데이터 시각화 입문/ASOS Data')

# 파일 경로 저장하기
data_scr = './drive/MyDrive/Colab Notebooks/알파코 파이썬 데이터 시각화 입문/ASOS Data/'

# 데이터 가져와서 합치기
df = pd.DataFrame()
for i in file_list:
    file_src = data_scr+i
    data = pd.read_csv(file_src, encoding='cp949')
    df = pd.concat([df,data], ignore_index=True) # 인덱스 재지정
df
```

## 2. 서울시 폭염/열대야 현황 시각화

### 2.1. 서울시 데이터 추출하여 서브셋 생성

```python
df_seoul = df[df['지점명']=='서울'].copy()
df_seoul
```

## 2.2. 서울시 폭염 데이터 추출
* 일 최고기온이 33도 이상

```python
df_seoul_hotday = df_seoul[df_seoul['최고기온(°C)']>=33].copy()
df_seoul_hotday
```

### 2.3. 서울시 폭염 데이터 시각화

```python
plt.rcParams['figure.figsize']=(15,5)
plt.plot(df_seoul_hotday['일시'], df_seoul_hotday['최고기온(°C)'], 'r.')
plt.xlabel('일시')
plt.ylabel('일 최고기온(°C)')
plt.title('서울시 폭염일 현황(1911~2020)',size=20)
plt.show()

# 계속해서 폭염일이 늘어나고 있다고는 볼 수 없지만 
# 최근 매우 더운날이 있었다.
```

### 2.4. 서울시 열대야 데이터 추출
* 일 최저기온 25도 이상

```python
df_seoul_hotnight = df_seoul[df_seoul['최저기온(°C)']>=25].copy()
df_seoul_hotnight
```

### 2.5. 서울시 열대야 데이터 시각화

```python
plt.plot(df_seoul_hotnight['일시'],df_seoul_hotnight['최저기온(°C)'],'k.')
plt.title('서울시 열대야 현황(1911~2020)', size=29)
plt.xlabel('일시')
plt.ylabel('일 최저기온(°C)')

plt.show()

# 열대야는 점차 늘어나는 추세이다.
```

## 3. 최고 폭염일, 최고 열대야 강조하기

### 3.1. 폭염이 가장 심했던 날

#### 3.1.1. 폭염이 가장 심했던 날 best10
* 데이터프레임.nlargest(개수,컬럼명) : 컬럼의 값이 큰 순으로 개수만큼 출력

```python
df_seoul_hotestday = df_seoul_hotday.nlargest(10,'최고기온(°C)')
df_seoul_hotestday
# 2018년이 굉장히 더웠다.
```

#### 3.1.2. 폭염이 가장 심했던 날짜, 최고기온

```python
df_seoul_hotestday = df_seoul_hotestday.iloc[0]
print(df_seoul_hotestday[2])
print(df_seoul_hotestday[5])
```

```python
# 시간은 없앤다.
print(df_seoul_hotestday[2].date())
print(df_seoul_hotestday[5])
```

#### 3.1.3. 시각화

```python
# 폭염일 그래프
plt.plot(df_seoul_hotday['일시'], df_seoul_hotday['최고기온(°C)'], 'r.')
plt.xlabel('일시')
plt.ylabel('일 최고기온(°C)')
plt.title('서울시 폭염일 현황(1911~2020)',size=20)

# 가장 더웠던 날 표시
plt.plot(df_seoul_hotestday[2].date(), df_seoul_hotestday[5],'r^')

# 가장 더웠던 날, 온도 표시
plt.text(df_seoul_hotestday[2].date(), df_seoul_hotestday[5]+0.5,
         str(df_seoul_hotestday[2].date())+" ("+str(df_seoul_hotestday[5])+"도)",
         ha='center', size=12)

plt.yticks(range(33,42)) # 글자가 그래프 밖으로 빠져 나가니까 y축을 늘려줌

plt.show()
```

## 4. 연도별 폭염일수, 열대야일수 추출

### 4.1. 연도별 폭염일수 추출

```python
# 연도별 폭염일수 : 연도로 그룹핑하여 일수 카운트
df_seoul_hotday_count = df_seoul_hotday.groupby(df_seoul_hotday['일시'].dt.year)['일시'].count().to_frame()
df_seoul_hotday_count
```

```python
# 폭염일이 없는 연도를 포함시키기 위하여 1911~2020년의 연도 데이터프레임 생성
df_years = pd.DataFrame(range(1911,2021))
df_years.index = df_years[0] # 인덱스 연도로 바꾸기
df_years
```

```python
# 연도별 폭염일수 연도 데이터 프레임 합치기
df_seoul_hotday_count = pd.concat([df_seoul_hotday_count, df_years], axis=1)
# 인덱스를 기준으로 두 데이터프레임을 옆으로 붙이는것.
df_seoul_hotday_count
```

```python
# 불필요한 컬럼 삭제
df_seoul_hotday_count.drop(columns=[0], inplace=True)
```

```python
# 결측치 확인
df_seoul_hotday_count.isnull().sum()
# 13개 연도에는 폭염일이 없었다.
```

```python
# 결측치를 0으로 채우기.
df_seoul_hotday_count.fillna(0, inplace=True)
df_seoul_hotday_count
```

```python
# 컬럼명 변경하기
df_seoul_hotday_count.columns=['폭염일수']
```

```python
# 폭염일수 데이터를 정수형으로 바꾸기
df_seoul_hotday_count = df_seoul_hotday_count.astype('int64')
df_seoul_hotday_count
```

```python
# 폭염이 가장 많았던 해의 데이터 best10
df_seoul_hotday_count.nlargest(10,'폭염일수')
```

### 4.2. 연도별 열대야일수 추출

```python
# 연도별 열대야일수 : 연도로 그룹핑하여 일수 카운트
df_seoul_hotnight_count = df_seoul_hotnight.groupby(df_seoul_hotnight['일시'].dt.year)['일시'].count().to_frame()
df_seoul_hotnight_count
```

```python
# 열대야가 없는 연도를 포함시키기 위하여 1911~2020년의 연도 데이터프레임과 합치기
df_seoul_hotnight_count = pd.concat([df_seoul_hotnight_count,df_years], axis=1)
# 인덱스를 기준으로 두 데이터프레임을 옆으로 붙이는것.
df_seoul_hotnight_count
```

```python
# 불필요한 컬럼 삭제
df_seoul_hotnight_count.drop(columns=[0], inplace=True)
df_seoul_hotnight_count
```

```python
# 결측치 확인
df_seoul_hotnight_count.isnull().sum()
# 28개 연도에는 열대야가 없었다.
```

```python
# 결측치를 0으로 채우기.
df_seoul_hotnight_count.fillna(0, inplace=True)
df_seoul_hotnight_count
```

```python
# 폭염일수 데이터를 정수형으로 바꾸기
df_seoul_hotnight_count = df_seoul_hotnight_count.astype('int64')
df_seoul_hotnight_count
```

```python
# 컬럼명 변경하기
df_seoul_hotnight_count.columns=['열대야일수']
```

```python
# 열대야가 가장 많았던 해의 데이터 best10
df_seoul_hotnight_count.nlargest(10,'열대야일수')
```
