# 22. 시리즈와 데이터프레임 다루기

* 데이터프레임에서 열에 접근하기:

    df[[열1,열2]]

* 데이터프레임에서 행,열에 접근하기:

    loc: 이름으로 접근

    iloc: 순서로 접근

    loc[0]-> 행이름이 0 // iloc[0] -> 0번째 행

    loc[[0,20,200],['year','country]] -> 0,20,200 행의 year, country 열

    iloc[[0,20,200],[0,1]] -> 0,20,200 번째 행의 0,1 번째 열

* 내가 원하는 행만 추출하기: df[조건]
* 머리부분 추출:

    df.head(n=10) -> 5개가 디폴트

* 꼬리부분 추출:
    df.tail()
* 모양 확인:
    df.shape()
* 열 이름 확인:
    df.coulums
* 행 이름 확인:
    df.index
* 자료형 확인:
    df.dtypes() // object - 문자열
* 정보 확인:
    df.info() -> 자료형 포함 용량, 크기


* 시리즈, 데이터프레임 중복 데이터 삭제하기:

    s.drop_duplicates()

    df.drop_duplicates()

* 시리즈 특정 값 바꾸기:
    s.replace(값,변경값)

* 시리즈 랜덤 샘플 추출:
    s.sample(n)
* 시리즈 정렬하기:

    s.sort_values() -> 오름차

    s.sort_values(ascending=False) -> 내림차

    s.sort_index() -> 행번호로 정렬

    s.sort_index(ascending = False) -> 행번호로 내림차 정렬

* 시리즈를 데이터프레임화:
    s.to_frame()

* 시간 데이터 변환:

    pd.to_datetime() -> 문자열로 된 날짜 데이터를 시간 데이터로 변환

# 23. 데이터를 저장하고 불러오는 방법

* csv 파일 불러오기

    pd.read_csv("파일이름.csv", sep="구분자" , encoding="문자표현방식")
* tsv 파일 불러오기

    pd.read_csv("파일이름.csv", sep="\t" , encoding="문자표현방식")
* 엑셀 파일 불러오기

    pd.read_excel("파일이름.xlsx", sheet_name = "시트이름")
    -> 셀 병합이 돼있는 경우 에러
* pickle 파일 불러오기

    pd.read_pickle("파일이름.pickle") -> 저장하기 위함/ 다른 read함수처럼 눈으로 보지는 못함

* csv 파일로 저장하기

    df.to_csv("파일이름".csv, index=False) -> 인덱스 유무
* tsv 파일로 저장하기

    df.to_csv("파일이름".csv, sep="\t", index=False) 
* 엑셀 파일로 저장하기

    df.to_excel("파일이름".xlsx)
* pickle 파일로 저장하기
 
    df.to_pickle("파일이름".pickle) -> 속도가 빠름

# 24. 데이터 시각화가 필요한 이유

평균, 표준편차 등 수치만으로 봤을 때 같다고 같은 데이터가 아님

대표적인 시각화 라이브러리
* matplotlib -> 그래프 (커스터마이징 편리, 겹쳐서 그리기 가능) // 디자인 감각 필요, 난이도 있음
* seaborn -> 데이터 분석 연습용(연습용 데이터 + 가시화), 자동으로 쉽게 그려줌 // 커스터 마이징 어려움
* pandas -> 데이터 분석 툴 (가시화 함수 존재 - plot(), 간당한 그림 그릴 때 유용)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# 그림을 그리기 위한 큰 도화지(fig) 하나 준비
fig = plt.figure()

# 행과 열에 자리를 할당 - (2,2) 행렬에 몇번째 자리 (4분할)
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# ax1.그림함수() -> 그래프 그리기
# ax1.settitle() -> 그래프 제목
# fig.suptitle() -> 메인 제목
# fig.tight_layout() -> 글자 안겹치게
```

실습

```python
# 앤스컴 : 평균, 표준편차, 상관계수, 회귀계수가 모두 같은 4개의 데이터 집단

ans = sns.load_dataset("anscombe")
ans
```

```python
data1 = ans[ans['dataset']== "I"]
data2 = ans[ans['dataset']== "II"]
data3 = ans[ans['dataset']== "III"]
data4 = ans[ans['dataset']== "IV"]
data4
```

```python
print(data1.mean())
print(data2.mean())
print(data3.mean())
print(data4.mean())
```

```python
print(data1.std())
print(data2.std())
print(data3.std())
print(data4.std())
```

```python
# 도화지 준비
fig = plt.figure()

# 틀 만들기
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# 그래프 그리기
ax1.scatter(data1['x'], data1['y'])
ax2.scatter(data2['x'], data2['y'])
ax3.scatter(data3['x'], data3['y'])
ax4.scatter(data4['x'], data4['y'])

# 그래프 겹쳐 그리기
ax1.plot(data1['x'], data1['y'], 'red')
ax2.plot(data2['x'], data2['y'], 'red')
ax3.plot(data3['x'], data3['y'], 'red')
ax4.plot(data4['x'], data4['y'], 'red')

# 그래프 제목 설정
ax1.set_title("Data1")
ax2.set_title("Data2")
ax3.set_title("Data3")
ax4.set_title("Data4")

# 메인 제목 설정
fig.suptitle("Data")

# 글자 안겹치게
fig.tight_layout()
```

# 25. matplotlib 라이브러리를 활용해서 그래프 그리기

```python
tips = sns.load_dataset("tips")
tips
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(tips['total_bill'], bins =10) # 막대기 개수
ax.hist(tips['total_bill'], bins =20)
ax.hist(tips['total_bill'], bins =30)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(tips['total_bill'], tips['tip'], c=tips['size'], s=tips['size']*20, alpha=0.5) 
# size 마다 다른 색,크기 alpha = 투명도
ax.set_xlabel("TOTAL BILL")
ax.set_ylabel("TIP")
ax.set_title("SCATTER")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

male = tips[tips['sex']=='Male']
female = tips[tips['sex']=='Female']
ax.boxplot([male['total_bill'],female['total_bill']])
# 여러개면 []로 감싸기
```
