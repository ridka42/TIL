문자열 전용 함수: format

```python
# 10개의 공간을 할당 한 후 오른쪽 정렬

text = "{:>10}".format("Hi")
print(text)
```

```python
# 10개의 공간을 할당 한 후 가운데 정렬

text = "{:^10}".format("Hi")
print(text)
```

```python
# f -> 실수다 2 -> 소수점 두자리로 표현

text = "{:.2f}".format(3.141592)
print(text)
```
