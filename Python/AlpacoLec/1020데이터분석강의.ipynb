# 11. 반복문 WHILE문

조건문이 참인 동안에 while문 아래의 문장이 반복해서 수행

반복문 강제로 빠져나가기 : break

반복문의 조건문으로 돌아가기 : continue

```python
hit =0

while hit <10:
  hit = hit +1
  print("나무를 {}번 찍었습니다.".format(hit))
  if hit == 10:
    print("나무가 쓰려졌습니다.")
```

```python
coffee = 10

while True:
  coffee = coffee -1
  print("커피를 팔고, {}잔 남았습니다.".format(coffee))
  if coffee ==0:
    print("매진되었습니다.")
    break
```

```python
a = 0

while a< 10:
  a += 1
  if a % 2 ==0:
    continue
  print(a)
```

while문을 사용해서 1부터 1000까지 3의 배수의 합을 구해보세요.

while문을 사용해서 1부터 100까지 더하는 코드를 작성 후, 결과값을 화면에 출력해보세요.

```python
# 1~1000 중 3의 배수 합
a=0
b=0
while a<1000:
  a+=1
  if a%3 ==0:
    b=a+b
print(b)
```

```python
# 1~100 까지 더하기

a=0
b=0
while a<100:
  a+=1
  b+=a
print(b)

# a<=100 이라고 하면 a+=1 때문에 101까지 계산됨
```

로또 당첨 번호가 총 몇 개의 조합이 있는 지 for 문을 사용해서 로또 당첨 번호의 총 개수를 구해보세요.

```python
count = 0
for n1 in range(1,46):
  for n2 in range(n1+1,46):
    for n3 in range(n2+1,46):
      for n4 in range(n3+1,46):
        for n5 in range(n4+1,46):
          for n6 in range(n5+1,46):
            count+=1
print(count)
# 중복 없는 순서 없는 조합
45*44*43*42*41*40/(6*5*4*3*2)

1/count
```

# 13. 함수

어떤 일을 수행한 다음에 그 결과를 내어놓는 것

```python
 # 입력 값이 없는 함수
def say():
  return 'Hi'
say()
```

```python
# 반환 값이 없는 함수
def add(a,b): # 입력 값은 존재 (매개변수)
  print("%d, %d의 합은 %d입니다."%(a,b,a+b))

a= add(4,5) # 함수가 실행되면서 print됨
print(a) # a 는 print만 실행하고 값이 저장되지 않음
```

```python
# 입력 값도 반환 값도 없는 함수
def say():
  print("Hi")
```

```python
# 입력 값이 몇 개가 될지 모를 때
def add(num1,*num):
  result = 0
  for i in num:
    result = result+i
    message = "{}개의 합은 {}입니다.".format(num1,result)
  return message

add(3,3,5,7)
```

```python
# 키워드 파라미터-> dictionary 형태로 저장
def intro (**name):
  print(name)

intro(name='minsol',age=27) # key 값 넣을떄 '' 안씀
```

```python
# 함수를 빠져 나갈 때
def say_nick(nick):
  if nick == 'fool':
    return
  print("my nickname is {}".format(nick))

say_nick('apple')
say_nick('fool')
say_nick('banana')
```

```python
# 기본값 설정
def say_myname(name, old, man=True):
  print("my name is {}.".format(name))
  print("my age is {}.".format(old))
  if man == True:
    print("I'm man")
  else:
    print("I'm woman")

say_myname('minsol',27,False)
```

재귀 함수

```python
def fac(n):
  a=1
  for i in range(1,n+1):
    a= a*i
  return a

fac(3)
```

```python
# 재귀 함수 사용

def fac(n):
  if n ==1 :
    return 1
  else:
    return n *fac(n-1) # fac(4) = 4 * fac(3) = 4* 3* fac(2) = 4* 3* 2*fac(1)

fac(4)
```

피보나치 수열에서 위치에 해당하는 값을 반환해주는 함수를 완성해보세요.

```python
def fibo(n):
  a = 0
  if n==1:
    return 0
  elif n==2:
    return 1
  else:
    return fibo(n-2)+fibo(n-1)
fibo(7)
```
